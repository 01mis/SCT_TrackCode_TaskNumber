***Analyze traffic accident data to identify patterns related to road conditions, weather, and time of day. Visualize accident hotspots and contributing factors.***

**# === Road Conditions vs Severity ===**

***TASK4***

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("RTA Dataset.csv")

# Group by Road surface conditions and Accident severity
road_severity = df.groupby(['Road_surface_conditions', 'Accident_severity']).size().unstack().fillna(0)

# Plot
road_severity.plot(kind='bar', stacked=True, colormap='plasma', figsize=(10,6))
plt.title("Accident Severity by Road Surface Conditions")
plt.xlabel("Road Surface Conditions")
plt.ylabel("Number of Accidents")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

***# === Weather Conditions ===***

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("RTA Dataset.csv")

# Count weather condition values
weather_counts = df['Weather_conditions'].value_counts()

# Plot
plt.figure(figsize=(8,5))
plt.figure(figsize=(10, 6))
bars = plt.barh(weather_counts.index, weather_counts.values, color=plt.cm.coolwarm(range(len(weather_counts))))
plt.title("Accidents by Weather Conditions")
plt.xlabel("Number of Accidents")
plt.ylabel("Weather Condition")
plt.tight_layout()
plt.show()
```

***# === Top 10 Accident Causes ===***

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("RTA Dataset.csv")

# Get top 10 causes
top_causes = df['Cause_of_accident'].value_counts().head(10)

# Plot

plt.figure(figsize=(10, 6))
colors = plt.cm.plasma(range(len(top_causes)))

plt.bar(top_causes.index, top_causes.values, color=colors)
plt.title("Top 10 Causes of Accidents")
plt.xlabel("Number of Accidents")
plt.tight_layout()
plt.xticks(rotation=45)
plt.show()
```

***# === Accident Hotspot ===***

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("RTA Dataset.csv")

# Get top 10 accident-prone areas
hotspots = df['Area_accident_occured'].value_counts().head(10)

# Plot
plt.figure(figsize=(10,6))


colors = plt.cm.viridis(range(len(hotspots)))
plt.bar(hotspots.index, hotspots.values, color=colors)
plt.title("Top 10 Accident-Prone Areas")
plt.xlabel("Number of Accidents")
plt.tight_layout()
plt.xticks(rotation=45)
plt.show()
```
