This repository contains a bar chart and histogram created using matplotlib to visualize:
- Distribution of Age (Histogram)
- Distribution of Departments (Bar chart)

```python
import pandas as pd

# Sample dataset with continuous variables like Age and Experience
data = pd.DataFrame({
    'EmployeeID': range(1, 21),
    'Age': [23, 27, 35, 29, 41, 38, 45, 26, 30, 34, 40, 31, 33, 37, 44, 28, 25, 39, 36, 32],
    'Experience': [1, 3, 10, 5, 18, 15, 20, 2, 6, 9, 17, 7, 8, 14, 19, 4, 2, 16, 13, 11],
    'Department': ['HR', 'Finance', 'IT', 'Marketing', 'IT', 'HR', 'Finance', 'Marketing', 'IT', 'HR',
                   'Finance', 'IT', 'Marketing', 'HR', 'Finance', 'Marketing', 'IT', 'Finance', 'HR', 'IT']
})

# Save to CSV (optional if you're working locally)
data.to_csv('sample_task1_dataset.csv', index=False)

print(data.head())
```

**TASK1**

```python
import matplotlib.pyplot as plt

plt.hist(data['Age'], bins=6, color='steelblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample dataset with continuous variables like Age and Experience
data = pd.DataFrame({
    'EmployeeID': range(1, 21),
    'Age': [23, 27, 35, 29, 41, 38, 45, 26, 30, 34, 40, 31, 33, 37, 44, 28, 25, 39, 36, 32],
    'Experience': [1, 3, 10, 5, 18, 15, 20, 2, 6, 9, 17, 7, 8, 14, 19, 4, 2, 16, 13, 11],
    'Department': ['HR', 'Finance', 'IT', 'Marketing', 'IT', 'HR', 'Finance', 'Marketing', 'IT', 'HR',
                   'Finance', 'IT', 'Marketing', 'HR', 'Finance', 'Marketing', 'IT', 'Finance', 'HR', 'IT']
})
dept_counts = data['Department'].value_counts()

plt.bar(dept_counts.index, dept_counts.values, color='coral')
plt.title('Department Distribution')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```
